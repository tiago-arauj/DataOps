```python
# Frameworks

# lib Data e hora 
import datetime 

# lib sistema operacional
import os 
import platform
import win32com.client as win32

# lib conexão
import requests
import socket 

#lib dados 
import pandas as pd

#lib visualizaçõa 
import matplotlib.pyplot as plt

#lib SQL
import sqlite3

# Lib para API Yahoo finance
import yfinance as yf
import pandas_datareader.data as web

# Avisos
import warnings
warnings.filterwarnings('ignore')
```

```python
# Rotinas iniciais

# Nome Rotinas
Id_Rotina = 1
Nome_Rotina = 'Fluxo de dados - booking'

# Indentificando o usuário

def Identificando_Usuario():
    
    #Capturando user
    Usuario = os.environ.get('USERNAME')

    #Capturando maquina
    Maquina = platform.node()

    #Capturando SO
    Sistema_Operacional = platform.platform()
    
    #Retorno da Função
    return( Usuario, Maquina, Sistema_Operacional )

#Identificar o inicio da execução
def Inicio_Rotina():
    
    #Data de Inicio
    Data_Inicio = datetime.datetime.today().date()
    
    #Hora de Inicio
    Hora_Inicio = datetime.datetime.now()
    
    #Retorno da Função
    return ( Data_Inicio, Hora_Inicio )

#Identificar o termino da execução
def Termino_Rotina():
    
    #Data de Inicio
    Data_Fim = datetime.datetime.today().date()
    
    #Hora de Inicio
    Hora_Fim = datetime.datetime.now()
    
    #Retorno da Função
    return ( Data_Fim, Hora_Fim )

# Verificando conexao de internet
def Verificar_Conexao():

    #Conexão
    URL = 'https://www.google.com'

    #tempo de conexão
    Timeout = 5

    try:
    #função para
        requests.get ( URL, timeout=TimeOut )
        return True

    except:
        return False
    
#Identificando IP
def Identificando_IP():
    
    try: 
        # Identificando IP Local
        IP_Local = socket.gethostbyname( socket.gethostname() )
        return IP_Local
    
    except:
        return False

#Gerando os parametros 
Lista_Usuario = Identificando_Usuario()
Lista_Inicio = Inicio_Rotina()
Lista_Termino = Termino_Rotina()
Conexao_Internet = Verificar_Conexao()
Verificando_IP = Identificando_IP()

# Erros operacionais
Erro_Operacional = ['']

print(
Lista_Usuario, Lista_Inicio,  Lista_Termino, Conexao_Internet, Verificando_IP, Erro_Operacional )
```

```python
# Operacional
try:
    
    # Pegando os dados
    Base_Dados = web.get_data_yahoo('VALE3', start='01-01-2022')
    Eixo_Plot = Base_Dados['Close']
    
    # Plot Gráfico
    
    # Tamanho
    plt.figure( figsize=(15, 6) )
    
    # Estilo
    plt.style.use('ggplot')
    
    # Plot do gráfico
    plt.plot( Eixo_Plot )
    
    # Labels, Titulos
    plt.title('Análise das ações')
    plt.xlabel('Periodo')
    plt.ylabel('Preço de Fechamento')
    
    # Exportar
    plt.savefig('Indicador_Ações.png')
    
except:
    Erro_Operacional = 'Erro_Execução'
```

```python
# Termino Rotina 
Lista_Fim = Termino_Rotina()

#Organização
Dicionario = {
    'Id_Rotina' : Id_Rotina,
    'Nome_Rotina' : Nome_Rotina,
    'Usuario' : Lista_Usuario[0],
    'Maquina' : Lista_Usuario[1],
    'Sistema_Operacional' : Lista_Usuario[2],
    'Data_Inicio' : Lista_Inicio[0],
    'Horario_Inicio' : Lista_Inicio[1],
    'Teste_Conexao' : Conexao_Internet,
    'IP_Local' : Verificando_IP,
    'Data_Termino' : Lista_Termino[0],
    'Horario_Termino' : Lista_Termino[1],
    'Tempo_Execucao' : ( Lista_Termino[1] - Lista_Inicio[1] ),
    'Erro' : Erro_Operacional
}

#Tabela de Logs 
Tabela_Log = pd.DataFrame( Dicionario, index=[0] )

# ---- Conexao SQL

# Criar a conexão
Conexao = sqlite3.connect('Banco_Dados.db')

# Apontar
Cursor = Conexao.cursor()

# Enviando dados
Tabela_Log.to_sql(
    
    # Nome Tabela
    'Tabela_Processamento',
    
    # Conexao
    Conexao,
    
    # Se a tabela existe
    if_exists='append',
    
    # Ignorar index
    index=False
    
)

#Verificando...
Tabela_Log
```
